```python
from pathlib import Path
import pandas as pd
from bs4 import BeautifulSoup

# HTML ファイルのパス（さっき確認したもの）
base_dir = Path.cwd().parent
html_path = base_dir / "data" / "raw" / "jra" / "race_pw01sde1008202503091120251026_59.html"

# HTML を読み込んで BeautifulSoup でパース
html = html_path.read_text(encoding="utf-8")
soup = BeautifulSoup(html, "lxml")

# 「着順」ヘッダを持つ table を探して、それだけを pandas に渡す
result_table = soup.find("th", class_="place").find_parent("table")

df_raw = pd.read_html(str(result_table))[0]
df_raw.head()
```

```python
df = df_raw.copy()

# 列名を英語ベースにリネーム（機械学習用）
df = df.rename(columns={
    "着順": "finish_rank",
    "枠": "bracket_no",
    "馬番": "horse_no",
    "馬名": "horse_name",
    "性齢": "sex_age",
    "負担重量": "carried_weight",
    "騎手名": "jockey_name",
    "タイム": "time_str",
    "差": "margin_str",              # 「差」になっている場合
    "着差": "margin_str",           # 「着差」になっている場合の保険
    "コーナー通過順位": "corner_order",
    "推定上り": "last_3f_est",
    "馬体重（増減）": "body_weight_raw",
    "馬体重(増減)": "body_weight_raw",
    "調教師名": "trainer_name",
    "単勝人気": "popularity",
    "人気": "popularity",           # 列名が「人気」の場合のため
    "Rt": "rating",
})

df.head()
```

```python
# 性齢例: "牡3", "牝4", "セ5" など
df[["sex", "age"]] = df["sex_age"].str.extract(r"([牡牝セ騙])(\d+)")
df["age"] = df["age"].astype("Int64")
```

```python
# ---------------------------------------------------------
# 馬体重「456(-12)」形式 → 本体重・増減 を分解するセル（完全版）
# ---------------------------------------------------------

import pandas as pd

# 1) 「馬体重（増減）」の列を自動検出してコピー
weight_col = [c for c in df.columns if "馬体重" in str(c)][0]

df["body_weight_raw"] = df[weight_col].astype(str)

# 2) 正規表現で「本体重」「増減」を抽出
#   全角マイナス「－」にも対応
pattern = r"(\d+)\s*\(([+\-−]?\d+)\)"

w = df["body_weight_raw"].str.extract(pattern)

# 3) 数値型(Int64)に変換（coerceで NaN 許容）
df["body_weight"] = pd.to_numeric(w[0], errors="coerce").astype("Int64")
df["body_weight_diff"] = pd.to_numeric(w[1], errors="coerce").astype("Int64")

# 4) 確認用
df[["body_weight_raw", "body_weight", "body_weight_diff"]].head()
```

```python
# === コーナー通過順位を corner_1〜corner_4 に分解する ===

import pandas as pd

# 1) 「コーナー通過順位」の元列名を自動で特定
corner_src = [c for c in df.columns if "コーナー" in str(c)][0]
print("コーナー通過順位の元列:", corner_src)

# 2) 文字列をスペース区切りで4つに分割
corners = (
    df[corner_src]
      .astype(str)
      .str.replace("\u3000", " ", regex=False)  # 全角スペース → 半角
      .str.strip()
      .str.split(r"\s+", expand=True)          # 空白で分割
      .iloc[:, :4]                              # 念のため4列だけに制限
)

corners.columns = ["corner_1", "corner_2", "corner_3", "corner_4"]

# 3) 数値型(Int64)に変換（欠損は NaN 許容）
for c in corners.columns:
    corners[c] = pd.to_numeric(corners[c], errors="coerce").astype("Int64")

# 4) 元の df に結合
df = pd.concat([df, corners], axis=1)

# 5) 確認（列名を修正）
df[["horse_name", corner_src, "corner_1", "corner_2", "corner_3", "corner_4"]].head()
```

```python
# タイム情報テーブル (<div class="result_time_data"> 配下)
time_table = soup.select_one("div.result_time_data table")
df_time = pd.read_html(str(time_table))[0]
df_time
```

```python
# ハロンタイム
lap_str = df_time.loc[df_time[0] == "ハロンタイム", 1].iloc[0]
lap_times = [float(x.strip()) for x in lap_str.split("-")]

# 上り
up_str = df_time.loc[df_time[0] == "上り", 1].iloc[0]
# 例: "4F 47.4 - 3F 35.3" → 4F,3F のタイムだけ取り出す
import re
m = re.findall(r"(\d+)F\s+([\d\.]+)", up_str)
up_dict = {f"{k}F": float(v) for k, v in m}

lap_times, up_dict
```
